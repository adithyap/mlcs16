```python
def remove_stuff_from_citations(cite): #for citation, priorpub columns.
    import re
    citelist = re.findall('[a-zA-Z/]+|\\d+', cite.lower()) #remove punctuation,spaces
    citelist = [s.lstrip("0") for s in citelist] #remove insignificant 0s
    cite2 = [''.join(row) if 'z' not in row else ''.join(row[:-1]) for row in [citelist]][0] #remove trailing z or zz's
    if cite2[-1]=='d': #append 0 if page starts at 0
        cite2 = cite2+'0'
    return cite2

def remove_stuff_from_citecol(cite): #only for cite column, as it was giving trouble.
    import re
    citelist = re.findall('[/]+|\\d+', repr(cite).lower()) #remove punctuation,spaces
    citelist = [s.lstrip("0") for s in citelist] #remove insignificant 0s
    cite2 = [''.join(row) if 'z' not in row else ''.join(row[:-1]) for row in [citelist]][0] #remove trailing z or zz's
#     if cite2[-1]=='d': #append 0 if page starts at 0
#         cite2 = cite2+'0'
    return cite2
```

There are about 36 rows out of 732 which have a '/' instead of f2d/f3d. We have to replace the '/' with a f2d/f3d depending on the year of citation etc.

```python
filtered_merged_caselevel['cite']=filtered_merged_caselevel['cite'].apply(remove_stuff_from_citecol)
```

```python
#to replace '/' in priorpub with 0 or respective prior entries
def clean_citations(df,colname):
    """
    Replaces '/' in df[colname] with f2d or f3d depending on 
    case year, and f3d/f2d year. Expects df to have a 'year'& 'cite' column.
    """
    slash='/'
    citelist = df['cite']
    for ind,i in enumerate(df[colname]):
        if slash in i:
            if i in citelist: #if priorpub entry in cite
                citeind=citelist.index(i)
                newcite = df['citation'][citeind]
                df[colname][ind]=newcite
            else:
                df[colname][ind]= 0
    return df
```
