```python
import pandas as pd
vot=pd.read_stata('Votelevel_stuffjan2013.dta')
print vot.shape
```

```python
# here I try to make a sparse dataframe to represent each judge's alignment, now just the first 10th cases
record=[]
for i in range(10):
    temp=dict()
    temp['casenum']=vot.ix[i][0]     #record case id
    temp[vot.ix[i][383]]=1           #first judge alignment, always 1 
    if vot.ix[i][159]==1:           #second judge alignment
        temp[vot.ix[i][156]]=1
    else:               
        temp[vot.ix[i][156]]=-1
    if vot.ix[i][164]==1:           #third judge alignment
        temp[vot.ix[i][161]]=1
    else:
        temp[vot.ix[i][161]]=-1
    record.append(temp)
    
#print record
testlist=pd.DataFrame(record)
```
